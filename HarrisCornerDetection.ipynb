# Harris Corner Detection

### Import resources and display image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
imag = cv2.imread('images/chess.jpg')

image_copy = np.copy(imag)

# Change color to RGB (from BGR)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

### Corners Detection

```python
gray = cv2.cvtColor(image_copy, cv2.COLOR_RGB2GRAY)

gray = np.float32(gray)

# Detect corner

corner = cv2.cornerHarris(gray, 2,3,0.03)

corner = cv2.dilate(corner,None)
plt.imshow(corner, cmap='gray')
```

### Display strong corners

```python
threshold = 0.001*corner.max()

orginal_image = np.copy(image_copy)


for i in range(0, corner.shape[0]):
    for j in range(0, corner.shape[1]):
        if (corner[i,j] > threshold):
            cv2.circle(orginal_image, (i,j),1,(255,0,0),10)
            
plt.imshow(orginal_image)
```
