# OpenCv

```python
imag = cv2.imread("images/cat.jpg")
print("Dimension:",imag.shape)
```

```python

plt.imshow(imag)
```

```python
imag_copy = np.copy(imag)

imag_copy = cv2.cvtColor(imag_copy, cv2.COLOR_BGR2RGB)

plt.imshow(imag_copy)
```

# Color threshold

```python
lower_blue = np.array([0,0,150])
upper_blue = np.array([80,80,255])
mask = cv2.inRange(imag_copy,lower_blue,upper_blue)
plt.imshow(mask,cmap='gray')
```

```python
mask_img = np.copy(imag_copy)
mask_img[mask!=0]=[0,0,0]
plt.imshow(mask_img)
```

```python
background_img = cv2.imread("images/space1.jpg")
plt.imshow(background_img)
```

```python
background_img = cv2.cvtColor(background_img, cv2.COLOR_BGR2RGB)
background_img[mask == 0]=[0, 0, 0]
plt.imshow(background_img)
```

```python
complete_img = background_img + mask_img
plt.imshow(complete_img)
```

```python
ball = cv2.imread("images/ball.jpg")
plt.imshow(ball)
```

```python
ball_copy= np.copy(ball)
ball_copy = cv2.cvtColor(ball_copy, cv2.COLOR_BGR2RGB)
plt.imshow(ball_copy)
```

## RGB channel

```python
r = ball_copy[:,:,0]
g = ball_copy[:,:,1]
b = ball_copy[:,:,2]

f,(ax1,ax2,ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title("Red")
ax1.imshow(r, cmap='gray')

ax2.set_title("Green")
ax2.imshow(g,cmap='gray')

ax3.set_title("Blue")
ax3.imshow(b, cmap='gray')
```

# HSV channel

```python
hsv = cv2.cvtColor(ball_copy, cv2.COLOR_RGB2HSV)

h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

f,(ax1,ax2,ax3) = plt.subplots(1, 3, figsize=(20,10))

ax1.set_title("Red")
ax1.imshow(h, cmap='gray')

ax2.set_title("Green")
ax2.imshow(s,cmap='gray')

ax3.set_title("Blue")
ax3.imshow(v, cmap='gray')
```

# RGB thresshold

```python
lower_pink = np.array([180,0,100])
upper_pink = np.array([255,255,230])
```

```python
lower_hue = np.array([160,0,0])
upper_hue = np.array([180,255,255])
```

```python
mask_rgb = cv2.inRange(ball_copy,lower_pink, upper_pink)
mask_ball = np.copy(ball_copy)
mask_ball[mask_rgb ==0]=[0,0,0]
plt.imshow(mask_ball)
plt.title("RGB Selection")
```

```python
mask_hsv = cv2.inRange(hsv,lower_hue, upper_hue)
mask_ball = np.copy(ball_copy)
mask_ball[mask_hsv ==0]=[0,0,0]
plt.imshow(mask_ball)
plt.title("HSV Selection")
```
